# Data Cleaning Part 3: Supplementary files
---

In this section, we are going to work on two supplementary files having data on property types and addresses. Remove the duplicated keys('ACCOUNT_NUM') and sorting out relevant columns for our analysis.

```python
import pandas as pd
import numpy as np
```

```python
# Read from the csv files
df_info = pd.read_csv('./Resources/project1_account_info.csv')
df_res_d = pd.read_csv('./Resources/project1_res_detail.csv')
```

```python
# Detecting duplicated keys
df_res_d['ACCOUNT_NUM'].value_counts()
```

```python
# Detecting duplicated keys
df_info['ACCOUNT_NUM'].value_counts()
```

```python
df_res = pd.read_csv('./Resources/res_detail.csv')
df_res
```

```python
df_res['ACCOUNT_NUM'].value_counts()
```

```python
df_res = df_res.drop_duplicates(subset='ACCOUNT_NUM', keep='last')
```

```python
df_res['ACCOUNT_NUM'].value_counts()
```

```python
df_res_d.columns
```

```python
df_res = df_res[df_res_d.columns]
```

```python
df_res
```

```python
df_res['ACCOUNT_NUM'].value_counts()
```

```python
df_res.to_csv('./Resources/res_detail_new.csv', index=False)
```
